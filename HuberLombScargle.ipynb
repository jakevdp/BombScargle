# Robust Lomb-Scargle

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
# Create some data
np.random.seed(42)
t = 10 * np.random.rand(100)
dy = 0.1 + 0.1 * np.random.rand(100)
y = np.random.normal(np.sin(t) + 0.5 * np.sin(2 * t), dy)

# Add some outliers
i = np.random.random(len(y))
mask = (i < 0.2)
y[mask] += 2 * np.random.randn(mask.sum())
```

```python
# Plot the data
plt.errorbar(t, y, dy, fmt='.k', ecolor='gray');
```

```python
from bombscargle.standard import lomb_scargle, lomb_scargle_huber

omega = np.logspace(-1, 1, 500)

# Standard Lomb-Scargle result
PLS = lomb_scargle(t, y, dy, omega, Nterms=2,
                   compute_offset=True)

# Huber Lomb-Scargle result
PLS_huber = lomb_scargle_huber(t, y, dy, omega, Nterms=2, c=5,
                               compute_offset=True)
```

```python
plt.plot(omega, PLS, label='standard')
plt.plot(omega, PLS_huber, label='huber')
plt.ylim(0, 1)
plt.legend(frameon=False);
```
